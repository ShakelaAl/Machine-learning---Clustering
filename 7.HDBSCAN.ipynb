```python
import hdbscan
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

# Example dataset creation
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Define a range of min_cluster_size and min_samples values to test
min_cluster_size_values = range(5, 51, 5)
min_samples_values = range(1, 21, 2)

# Initialize lists to store results
min_cluster_size_list = []
min_samples_list = []
num_clusters_list = []

# Perform HDBSCAN for each combination of min_cluster_size and min_samples
for min_cluster_size in min_cluster_size_values:
    for min_samples in min_samples_values:
        clusterer = hdbscan.HDBSCAN(min_cluster_size=min_cluster_size, min_samples=min_samples)
        labels = clusterer.fit_predict(X)
        
        # Number of clusters found (excluding noise)
        num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        
        min_cluster_size_list.append(min_cluster_size)
        min_samples_list.append(min_samples)
        num_clusters_list.append(num_clusters)
        
        # Optional: Print or store the number of clusters for each min_cluster_size and min_samples
        print(f'min_cluster_size: {min_cluster_size}, min_samples: {min_samples} -> Number of clusters: {num_clusters}')

# Plotting the number of clusters for each combination of min_cluster_size and min_samples
plt.figure(figsize=(12, 8))
plt.scatter(min_cluster_size_list, min_samples_list, c=num_clusters_list, cmap='viridis', s=50, edgecolor='k')
plt.colorbar(label='Number of Clusters')
plt.title('HDBSCAN: Number of Clusters vs. min_cluster_size and min_samples')
plt.xlabel('min_cluster_size')
plt.ylabel('min_samples')
plt.grid(True)
plt.show()

# Use a specific combination of min_cluster_size and min_samples (for example, min_cluster_size=15 and min_samples=5)
best_min_cluster_size = 15
best_min_samples = 5
clusterer = hdbscan.HDBSCAN(min_cluster_size=best_min_cluster_size, min_samples=best_min_samples)
labels = clusterer.fit_predict(X)

# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Plot clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, s=50, cmap='viridis', edgecolor='k')
plt.title(f'HDBSCAN Clustering with min_cluster_size={best_min_cluster_size} and min_samples={best_min_samples}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
