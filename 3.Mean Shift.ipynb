```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt

# Initialize Mean Shift with default parameters
mean_shift = MeanShift()

# Fit the Mean Shift model
mean_shift.fit(X)

# Get the cluster centers and labels
cluster_centers = mean_shift.cluster_centers_
labels = mean_shift.labels_

# Plotting (if you want to visualize cluster centers)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Mean Shift Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Optionally, print the number of clusters found
num_clusters = len(set(labels))
print(f'Number of clusters found: {num_clusters}')
```

```python
from sklearn.cluster import MeanShift
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Define a range of bandwidth values to test
bandwidth_values = np.linspace(0.1, 2.0, 10)

# Initialize lists to store results
num_clusters_list = []

# Perform Mean Shift clustering for each bandwidth value
for bandwidth in bandwidth_values:
    mean_shift = MeanShift(bandwidth=bandwidth)
    mean_shift.fit(X)
    
    # Number of clusters found
    num_clusters = len(set(mean_shift.labels_))
    num_clusters_list.append(num_clusters)
    
    # Optional: Print or store the number of clusters for each bandwidth
    print(f'Bandwidth: {bandwidth:.2f} -> Number of clusters: {num_clusters}')

# Plotting the number of clusters found for each bandwidth
plt.plot(bandwidth_values, num_clusters_list, marker='o')
plt.title('Number of Clusters vs. Bandwidth')
plt.xlabel('Bandwidth')
plt.ylabel('Number of Clusters')
plt.grid(True)
plt.show()

# Use the best bandwidth found (for example, the first one)
best_bandwidth = bandwidth_values[0]
mean_shift = MeanShift(bandwidth=best_bandwidth)
mean_shift.fit(X)

# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Get the cluster centers and labels
cluster_centers = mean_shift.cluster_centers_
labels = mean_shift.labels_

# Reduce the cluster centers to 2D
cluster_centers_reduced = pca.transform(cluster_centers)

# Plot clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, s=50, cmap='viridis', label='Data Points')
plt.scatter(cluster_centers_reduced[:, 0], cluster_centers_reduced[:, 1], c='red', s=200, alpha=0.75, marker='x', label='Cluster Centers')
plt.title('Mean Shift Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
