```python
import numpy as np
from sklearn.cluster import AffinityPropagation
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load dataset (example with Iris dataset)
data = load_iris()
X = data.data

# Create and fit Affinity Propagation model
aff_prop = AffinityPropagation()
aff_prop.fit(X)

# Get cluster centers and labels
cluster_centers_indices = aff_prop.cluster_centers_indices_
labels = aff_prop.labels_

# Number of clusters
n_clusters = len(cluster_centers_indices)

print(f'Number of clusters: {n_clusters}')
print(f'Cluster centers indices: {cluster_centers_indices}')
print(f'Labels: {labels}')

# Visualize the clusters (only for the first two features for simplicity)
plt.figure(figsize=(10, 7))

# Scatter plot for clusters
for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Cluster {i}')

# Plot cluster centers
plt.scatter(X[cluster_centers_indices, 0], X[cluster_centers_indices, 1], s=300, c='red', marker='X', label='Cluster Centers')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Affinity Propagation Clustering')
plt.legend()
plt.show()
```
