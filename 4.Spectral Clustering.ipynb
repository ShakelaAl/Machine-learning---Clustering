```python
from sklearn.cluster import SpectralClustering
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

# Example dataset creation
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Define a range of number of clusters to test
num_clusters_range = range(2, 11)  # Spectral clustering requires at least 2 clusters

# Initialize lists to store results
num_clusters_list = []

# Perform Spectral Clustering for each number of clusters
for n_clusters in num_clusters_range:
    spectral = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
    labels = spectral.fit_predict(X)
    
    # Number of clusters found
    num_clusters_list.append(n_clusters)
    
    # Optional: Print or store the number of clusters for each n_clusters
    print(f'Number of clusters: {n_clusters} -> Spectral Clustering complete')

# Plotting the number of clusters
plt.plot(num_clusters_range, num_clusters_list, marker='o')
plt.title('Spectral Clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('Number of Clusters')
plt.grid(True)
plt.show()

# Use a specific number of clusters (for example, 4)
best_n_clusters = 4
spectral = SpectralClustering(n_clusters=best_n_clusters, affinity='nearest_neighbors', random_state=42)
labels = spectral.fit_predict(X)

# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Plot clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, s=50, cmap='viridis')
plt.title(f'Spectral Clustering with {best_n_clusters} Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
