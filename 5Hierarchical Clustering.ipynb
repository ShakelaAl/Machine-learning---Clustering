```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

# Example dataset creation
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Define a range of n_clusters values to test
n_clusters_values = range(2, 10)

# Initialize lists to store results
n_clusters_list = []

# Perform Agglomerative Clustering for each n_clusters
for n_clusters in n_clusters_values:
    agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
    labels = agg_clustering.fit_predict(X)
    
    # Number of clusters found
    num_clusters = len(set(labels))
    
    n_clusters_list.append(num_clusters)
    
    # Optional: Print or store the number of clusters for each n_clusters
    print(f'n_clusters: {n_clusters} -> Number of clusters: {num_clusters}')

# Plotting the number of clusters for each n_clusters
plt.figure(figsize=(12, 8))
plt.plot(n_clusters_values, n_clusters_list, marker='o')
plt.title('Agglomerative Clustering: Number of Clusters vs. n_clusters')
plt.xlabel('n_clusters')
plt.ylabel('Number of Clusters')
plt.grid(True)
plt.show()

# Use a specific n_clusters value (for example, n_clusters=4)
best_n_clusters = 4
agg_clustering = AgglomerativeClustering(n_clusters=best_n_clusters)
labels = agg_clustering.fit_predict(X)

# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Plot clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, s=50, cmap='viridis', edgecolor='k')
plt.title(f'Agglomerative Clustering with n_clusters={best_n_clusters}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
